## Neural Network

```python
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
```

### Загружаем данные

```python
dataset = loadtxt('pima-indians-diabetes.csv', delimiter=',')
# разделили столбцы данных на входные и выходные переменные
X = dataset[:,0:8]
y = dataset[:,8]
```

```python
dataset.shape
```

Для каждого атрибута: (все числовые)
    входные(Х): 
   1. Количество беременностей
   2. Концентрация глюкозы в плазме через 2 часа в пероральном тесте на толерантность к глюкозе.
   3. Диастолическое артериальное давление (мм рт. Ст.)
   4. Толщина кожной складки трицепса (мм)
   5. 2-часовой сывороточный инсулин (мЕд / мл)
   6. Индекс массы тела (вес в кг / (рост в м) ^ 2)
   7. Диабетическая родословная.
   8. Возраст (лет)
   выходные(у): 
   9. Переменная класса (0 или 1)

```python
print(dataset)
```

### Определяем модель Keras

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

### Компиляция модели

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

### Обучение модели

```python
model.fit(X, y, epochs=150, batch_size=10)
```

```python
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
#для проверки изменения кол-ва эпох
model2 = Sequential()
model2.add(Dense(12, input_dim=8, activation='relu'))
model2.add(Dense(8, activation='relu'))
model2.add(Dense(1, activation='sigmoid'))
model2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model2.fit(X, y, epochs=50, batch_size=10)
```

```python
_, accuracy = model2.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```
